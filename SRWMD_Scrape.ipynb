```python
import pandas as pd
import numpy as np
import requests
import lxml.html as lh
from bs4 import BeautifulSoup
```

```python

srwmd_all_sites_df = pd.read_csv('C:/Users/robert.taylor/Documents/ArcGIS/srwmd_all_sites.txt',header=0,skiprows=[1,1],converters={'PopupInfo': lambda x: str(x)})
display(srwmd_all_sites_df)

srwmd_all_sites_df[['Active','station_nm','dec_lat_va','dec_long_va','data_start','data_end','stn_type','county','none']] = srwmd_all_sites_df['PopupInfo'].str.split('<br>',expand=True)

srwmd_all_sites_df['Active'] = srwmd_all_sites_df.Active.str.extract(r'Active-2019:\s(.*)',expand=False)
srwmd_all_sites_df['Active'] = srwmd_all_sites_df.loc[srwmd_all_sites_df['Active'] == ''].Active.str.replace('','No',regex=True)
srwmd_all_sites_df['Active'].fillna('Yes', inplace=True)
srwmd_all_sites_df['station_nm'] = srwmd_all_sites_df.station_nm.str.extract(r'Station_Name:\s(.*)',expand=False)
srwmd_all_sites_df['dec_lat_va'] = srwmd_all_sites_df.dec_lat_va.str.extract(r'LATITUDE:\s(.*)',expand=False)
srwmd_all_sites_df['dec_long_va'] = srwmd_all_sites_df.dec_long_va.str.extract(r'LONGITUDE:\s(.*)',expand=False)
srwmd_all_sites_df['data_start'] = srwmd_all_sites_df.data_start.str.extract(r'Data Starts:\s(.*)',expand=False)
srwmd_all_sites_df['data_end'] = srwmd_all_sites_df.data_end.str.extract(r'Latest Data:\s(.*)',expand=False)
srwmd_all_sites_df['stn_type'] = srwmd_all_sites_df.stn_type.str.extract(r'STNTYPE:\s(.*)',expand=False)
srwmd_all_sites_df['county'] = srwmd_all_sites_df.county.str.extract(r'County:\s(.*)',expand=False)
srwmd_all_sites_df.drop('PopupInfo', axis=1, inplace=True)
srwmd_all_sites_df.drop('none', axis=1, inplace=True)

srwmd_all_sites_df
srwmd_all_sites_df.to_csv('C:/Users/robert.taylor/Documents/SuwData/SRWMD_All_Sites_Fixed.csv')
```
