```python
count = 0
usgs_dict={}
    

for i in range(len(usgs_sites_retrieve)):
    usgs_site_no = usgs_sites_retrieve[i]
    url = 'https://waterdata.usgs.gov/nwis/dv?referred_module=sw&search_site_no='+usgs_site_no+'&search_site_no_match_type=exact&site_tp_cd=SP&dv_count_nu=20&index_pmcode_00055=1&index_pmcode_00060=1&index_pmcode_00061=1&index_pmcode_00065=1&index_pmcode_30207=1&index_pmcode_30208=1&index_pmcode_30209=1&index_pmcode_62611=1&index_pmcode_62613=1&index_pmcode_00010=1&index_pmcode_00011=1&index_pmcode_00600=1&index_pmcode_00613=1&index_pmcode_00618=1&index_pmcode_00620=1&index_pmcode_00630=1&index_pmcode_00631=1&group_key=NONE&sitefile_output_format=html_table&column_name=agency_cd&column_name=site_no&column_name=station_nm&column_name=site_tp_cd&column_name=dec_lat_va&column_name=dec_long_va&column_name=qw_begin_date&column_name=qw_end_date&column_name=qw_count_nu&range_selection=date_range&begin_date=1980-01-01&end_date=2021-08-03&format=rdb&date_format=YYYY-MM-DD&rdb_compression=value&list_of_search_criteria=search_site_no%2Csite_tp_cd%2Cobs_count_nu%2Crealtime_parameter_selection'
    page = urllib.request.urlopen(url)
    skiplines = str(page.read())
    skiprow = skiplines.count('#')
    usgs_site_data = pd.read_csv(url,sep='\t',skiprows=skiprow)
    usgs_site_data.drop([0],axis=0,inplace=True)
    
            
            
    
    usgs_dict['site_no_'+str(usgs_site_no)] = usgs_site_data
    count += 1
```
